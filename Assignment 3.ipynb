Question Set A

```python
# Q.1) Write a program to find the maximum and minimum element in a tuple without using max() and min() functions.

t = tuple(map(int, input("Enter elements separated by space: ").split()))

max_val = t[0]
min_val = t[0]

for i in t:
    if i > max_val:
        max_val = i
    if i < min_val:
        min_val = i

print("Maximum element:", max_val)
print("Minimum element:", min_val)
```

```python
# Q.2) Write a program to count the occurrence of each element in a tuple.
# Example: For tuple (1, 2, 3, 2, 1, 3, 2), output should show 1:2, 2:3, 3:2

t = (4, 6, 5, 6, 4, 5, 6)
count = {}

for i in t:
    if i in count:
        count[i] = count[i] + 1
    else:
        count[i] = 1

for key in count:
    print(key, ":", count[key])
```

```python
# Q.3) Write a program to convert a tuple of strings into a single string.
# Example: ("Hello", "World", "Python") should become "HelloWorldPython"

t = ("Just", "Do", "It")
result = ""

for word in t:
    result = result + word

print("Single string:", result)
```

```python
# Q.4) Write a program to find the sum of all even numbers in a tuple.

t = (1, 2, 3, 4, 5, 6, 8, 9, 10)
sum_even = 0

for i in t:
    if i % 2 == 0:
        sum_even = sum_even + i

print("Sum of even numbers:", sum_even)
```

```python
# Q.5) Write a program to remove duplicates from a tuple (return a new tuple with unique elements).

t = (1, 2, 3, 2, 4, 1, 5)
unique = []

for i in t:
    if i not in unique:
        unique.append(i)

new_tuple = tuple(unique)

print("Tuple without duplicates:", new_tuple)
```

Question Set B

```python
# Q.1) Write a program to find the second largest element in a tuple.

t = (10, 3, 14, 20, 8, 25)
lst = list(t)
lst.sort()

print("Second largest element:", lst[-2])
```

```python
# Q.2) Write a program to sort a tuple of tuples based on the second element.

t = ((1, 3), (4, 1), (2, 2), (5, 0))
lst = list(t)

for i in range(len(lst)):
    for j in range(i + 1, len(lst)):
        if lst[i][1] > lst[j][1]:
            lst[i], lst[j] = lst[j], lst[i]

sorted_tuple = tuple(lst)

print("Sorted tuple:", sorted_tuple)
```

```python
# Q.3) Write a program to check if all elements in a tuple are unique.

t = (1, 2, 3, 4, 5, 3)
s = set(t)

if len(t) == len(s):
    print("All elements are unique")
else:
    print("There are duplicates")
```

```python
# Q.4) Write a program to swap the first and last element of a tuple.

t = (10, 20, 30, 40, 50)
lst = list(t)

lst[0], lst[-1] = lst[-1], lst[0]

new_tuple = tuple(lst)

print("Tuple after swapping:", new_tuple)
```

```python
# Q.1) Write a program to find common elements between two tuples.
t1 = (1, 2, 3, 4, 5)
t2 = (3, 4, 5, 6, 7)
common = []

for i in t1:
    if i in t2:
        common.append(i)

common_tuple = tuple(common)

print("Common elements:", common_tuple)
```

```python
# Q.2) Write a program to create a tuple of squares of numbers from 1 to n.
# Example: If n=5, output should be (1, 4, 9, 16, 25)
n = int(input("Enter a number: "))
squares = []

for i in range(1, n+1):
    squares.append(i * i)

t = tuple(squares)

print("Tuple of squares:", t)
```
