```python
# Q.1 Write a program to create a dictionary with roll number as key and name as value for 5 students.
# Display all key-value pairs.

students = {}

for i in range(5):
    roll = input("Enter roll number: ")
    name = input("Enter name: ")
    students[roll] = name

print("Student dictionary:", students)
```

```python
# Q.2 Write a program to count the frequency of each word in a given sentence using a dictionary.
# Example: "hello world hello" â†’ {'hello': 2, 'world': 1}

sentence = input("Enter a sentence: ")
words = sentence.split()

freq = {}

for w in words:
    if w in freq:
        freq[w] += 1
    else:
        freq[w] = 1

print("Word frequency:", freq)
```

```python
# Q.3 Write a program to merge two dictionaries entered by the user.

dict1 = {}
n1 = int(input("How many elements in first dictionary? "))

for i in range(n1):
    key = input("Enter key: ")
    value = input("Enter value: ")
    dict1[key] = value

dict2 = {}
n2 = int(input("How many elements in second dictionary? "))

for i in range(n2):
    key = input("Enter key: ")
    value = input("Enter value: ")
    dict2[key] = value

dict1.update(dict2)

print("Merged dictionary:", dict1)
```

```python
# Q.4 Write a program to find the key with the maximum value in a dictionary

d = {}
n = int(input("Enter number of elements in dictionary: "))

for i in range(n):
    key = input("Enter key: ")
    value = int(input("Enter value: "))
    d[key] = value

max_key = list(d.keys())[0]

for key in d:
    if d[key] > d[max_key]:
        max_key = key

print("Key with maximum value:", max_key)
```

```python
# Q.5 Write a program to check if a given key exists in a dictionary

d = {'a': 10, 'b': 25, 'c': 15}
key = input("Enter key to check: ")

if key in d:
    print(key, "exists in the dictionary")
else:
    print(key, "does not exist in the dictionary")
```

Question Set B

```python
# Q.1 Write a program to remove a key from a dictionary and handle the case when key doesn't exist

d = {'a': 10, 'b': 25, 'c': 15}
key = input("Enter key to remove: ")

if key in d:
    d.pop(key)
    print("Updated dictionary:", d)
else:
    print("Key does not exist in the dictionary")
```

```python
# Q.2 Write a program to sort a dictionary by its values in descending order

d = {'a': 13, 'b': 5, 'c': 27}

sorted_d = dict(sorted(d.items(), key=lambda item: item[1], reverse=True))

print("Dictionary sorted by values (descending):", sorted_d)
```

```python
# Q.3 Write a program to create a dictionary from two lists

keys = input("Enter keys separated by space: ").split()
values = input("Enter values separated by space: ").split()

d = {}

for i in range(len(keys)):
    d[keys[i]] = values[i]

print("Dictionary:", d)
```

```python
# Q.1 Write a program to find all keys with the same value in a dictionary

d = {'a': 10, 'b': 20, 'c': 10, 'd': 30, 'e': 20}
value = int(input("Enter the value to search for: "))
keys_with_value = []

for key in d:
    if d[key] == value:
        keys_with_value.append(key)

if keys_with_value:
    print("Keys with value", value, ":", keys_with_value)
else:
    print("No key found with value", value)
```

```python
# Q.2 Write a program to create a dictionary where keys are numbers from 1 to n and values are their cubes

n = int(input("Enter a number: "))
d = {}

for i in range(1, n + 1):
    d[i] = i ** 3

print("Dictionary of cubes:", d)
```
