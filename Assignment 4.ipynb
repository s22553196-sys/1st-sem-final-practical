Question Set A

```python
# Q.1) Write a program to find the union, intersection, and difference of two sets entered by the user.
set1 = set(map(int, input("Enter elements of first set separated by space: ").split()))
set2 = set(map(int, input("Enter elements of second set separated by space: ").split()))

union_set = set1 | set2
intersection_set = set1 & set2
difference_set = set1 - set2

print("Union:", union_set)
print("Intersection:", intersection_set)
print("Difference (set1 - set2):", difference_set)
```

```python
# Q.2) Write a program to check if a set is a subset of another set.
set1 = set(map(int, input("Enter elements of first set: ").split()))
set2 = set(map(int, input("Enter elements of second set: ").split()))

if set1.issubset(set2):
    print("First set is a subset of second set")
else:
    print("First set is not a subset of second set")

if set2.issubset(set1):
    print("Second set is a subset of first set")
else:
    print("Second set is not a subset of first set")
```

```python
 # Q.3) Write a program to remove duplicates from a list using sets.
lst = list(map(int, input("Enter list elements separated by space: ").split()))
unique_set = set(lst)
unique_list = list(unique_set)

print("List without duplicates:", unique_list)
```

```python
# Q.4) Write a program to find elements that are present in the first set but not in the second set.
set1 = set(map(int, input("Enter elements of first set: ").split()))
set2 = set(map(int, input("Enter elements of second set: ").split()))

diff = set1 - set2

print("Elements in first set but not in second set:", diff)
```

```python
# Q.5) Write a program to check if two sets are disjoint (have no common elements).
set1 = set(map(int, input("Enter elements of first set: ").split()))
set2 = set(map(int, input("Enter elements of second set: ").split()))

if set1.isdisjoint(set2):
    print("Sets are disjoint")
else:
    print("Sets are not disjoint")
```

Question Set B

```python
# Q.1) Write a program to find symmetric difference between two sets.
set1 = set(map(int, input("Enter elements of first set: ").split()))
set2 = set(map(int, input("Enter elements of second set: ").split()))

sym_diff = set1 ^ set2

print("Symmetric difference:", sym_diff)
```

```python
# Q.2) Write a program to add multiple elements to a set at once from user input.
s = set(map(int, input("Enter initial set elements: ").split()))
new_elements = set(map(int, input("Enter elements to add: ").split()))
s.update(new_elements)

print("Updated set:", s)
```

```python
# Q.3) Write a program to find the length of a set and check if a specific element exists in it.
s = set(map(int, input("Enter set elements: ").split()))
x = int(input("Enter element to check: "))
length = len(s)

print("Length of set:", length)

if x in s:
    print(x, "exists in the set")
else:
    print(x, "does not exist in the set")
```

Question Set C

```python
# Q.1) Write a program that takes two sentences as input and finds:
# * Common words in both sentences
# * Words unique to the first sentence
# * Words unique to the second sentence

s1 = input("Enter first sentence: ")
s2 = input("Enter second sentence: ")

words1 = set(s1.split())
words2 = set(s2.split())

common = words1 & words2
unique1 = words1 - words2
unique2 = words2 - words1

print("Common words:", common)
print("Unique to first sentence:", unique1)
print("Unique to second sentence:", unique2)
```

```python
# Q.2) Write a program to perform all set operations (union, intersection, difference, symmetric difference) on three sets.
set1 = set(map(int, input("Enter elements of first set: ").split()))
set2 = set(map(int, input("Enter elements of second set: ").split()))
set3 = set(map(int, input("Enter elements of third set: ").split()))

union_set = set1 | set2 | set3
intersection_set = set1 & set2 & set3
difference_set = set1 - set2 - set3
sym_diff = (set1 ^ set2) ^ set3

print("Union:", union_set)
print("Intersection:", intersection_set)
print("Difference (set1 - set2 - set3):", difference_set)
print("Symmetric difference:", sym_diff)
```
