```python
#Q.1 Write a function to find the maximum of three numbers
def maximum(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

x = int(input("Enter first number: "))
y = int(input("Enter second number: "))
z = int(input("Enter third number: "))

print("Maximum number is:", maximum(x, y, z))
```

```python
#Q.2 Write a function to check if a number is prime or not
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

num = int(input("Enter a number: "))

if is_prime(num):
    print(num, "is a prime number")
else:
    print(num, "is not a prime number")
```

```python
#Q.3 Write a function to calculate factorial (iterative & recursive)
def factorial_iterative(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact

def factorial_recursive(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial_recursive(n - 1)

num = int(input("Enter a number: "))

print("Factorial (iterative):", factorial_iterative(num))
print("Factorial (recursive):", factorial_recursive(num))
```

```python
#Q.4 Write a function that accepts a list and returns a new list with unique elements
def unique_list(lst):
    unique = []
    for i in lst:
        if i not in unique:
            unique.append(i)
    return unique

lst = list(map(int, input("Enter list elements separated by space: ").split()))

print("List with unique elements:", unique_list(lst))
```

```python
#Q.5 Write a function to calculate the sum of all even numbers in a list
def sum_even(lst):
    total = 0
    for i in lst:
        if i % 2 == 0:
            total += i
    return total

lst = list(map(int, input("Enter list elements separated by space: ").split()))

print("Sum of even numbers:", sum_even(lst))
```

```python
#Q.1Write a function to convert temperature from Celsius to Fahrenheit and vice versa
def convert_temperature(temp, scale):
    if scale == "C":
        return (temp * 9/5) + 32
    elif scale == "F":
        return (temp - 32) * 5/9
    else:
        return None

temp = float(input("Enter temperature: "))
scale = input("Enter scale (C for Celsius, F for Fahrenheit): ").upper()

converted = convert_temperature(temp, scale)
if converted is not None:
    if scale == "C":
        print(f"{temp}°C = {converted:.2f}°F")
    else:
        print(f"{temp}°F = {converted:.2f}°C")
else:
    print("Invalid scale entered!")
```

```python
#Q.2 Write a function that accepts a string and returns the count of vowels and consonants
def count_vowels_consonants(s):
    vowels = "aeiouAEIOU"
    v_count = 0
    c_count = 0

    for char in s:
        if char.isalpha():
            if char in vowels:
                v_count += 1
            else:
                c_count += 1
    return v_count, c_count

text = input("Enter a string: ")
vowels, consonants = count_vowels_consonants(text)

print("Vowels:", vowels)
print("Consonants:", consonants)
```

```python
# Q.3) Write a function that accepts variable number of arguments and returns their average

def average(*args):
    total = 0
    count = len(args)
    for num in args:
        total += num
    return total / count if count != 0 else 0

nums = list(map(float, input("Enter numbers separated by space: ").split()))

print("Average is:", average(*nums))
```

```python
# Q.4) Write a lambda function to find the maximum of three numbers.

maximum = lambda a, b, c: a if (a >= b and a >= c) else (b if b >= c else c)

x = int(input("Enter first number: "))
y = int(input("Enter second number: "))
z = int(input("Enter third number: "))

print("Maximum number is:", maximum(x, y, z))
```

```python
# Q.1) Write a recursive function to calculate the sum of digits of a number.

def sum_of_digits(n):
    if n == 0:
        return 0
    else:
        return n % 10 + sum_of_digits(n // 10)

num = int(input("Enter a number: "))
print("Sum of digits:", sum_of_digits(num))
```

```python
# Q.2) Write a function that accepts a list of numbers and returns a dictionary with count of even and odd.
# Example: [1,2,3,4,5] should return {'even': 2, 'odd': 3}

def count_even_odd(lst):
    result = {'even': 0, 'odd': 0} # Dictionary to store counts
    for num in lst:
        if num % 2 == 0:
            result['even'] += 1
        else:
            result['odd'] += 1
    return result

numbers = list(map(int, input("Enter numbers separated by space: ").split()))

print("Even and Odd count:", count_even_odd(numbers))
```
