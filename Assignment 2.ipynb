```python
# Q.1) Write a program to count the number of vowels in a given string.
text = input("Enter a string: ")
vowels = "aeiouAEIOU"
count = 0

for ch in text:
    if ch in vowels:
        count += 1

print("Number of vowels:", count)
```

```python
# Q.2) Write a program to check if a string is a palindrome (reads the same forwards and backwards).
text = input("Enter a string: ")
reverse_text = text[::-1]

if text == reverse_text:
    print("Palindrome")
else:
    print("Not a Palindrome")
```

```python
# Q.3) Write a program to reverse a string without using built-in reverse functions.
text = input("Enter a string: ")
reversed_text = ""

for ch in text:
    reversed_text = ch + reversed_text

print("Reversed string is:", reversed_text)
```

```python
# Q.4) Write a program to count the frequency of each character in a string. 
# Example: For "hello", output should be h:1, e:1, l:2, o:1 
text = input("Enter a string: ")
freq = {}

for ch in text:
    if ch in freq:
        freq[ch] += 1
    else:
        freq[ch] = 1


for ch in freq:
    print(ch, ":", freq[ch], end=", ")
    
```

```python
# Q.5) Write a program to remove all whitespace from a string.

text = input("Enter a string: ")
result = ""

for ch in text:
    if ch != " ":          
        result += ch       

print("String without whitespace:", result)
```

```python
# Q.1) Write a program to find the first non-repeating character in a string.
text = input("Enter a string: ")
freq = {}

for ch in text:
    if ch in freq:
        freq[ch] += 1
    else:
        freq[ch] = 1

for ch in text:
    if freq[ch] == 1:
        print("First non-repeating character is:", ch)
        break
else:
    print("No non-repeating character found")
```

```python
# Q.2) Write a program to check if two strings are anagrams (contain same characters in different order)
s1 = input("Enter first string: ")
s2 = input("Enter second string: ")

flag = True
for ch in s1:
    if s1.count(ch) != s2.count(ch):
        flag = False
        break

if flag:
    print("Anagram")
else:
    print("Not Anagram")
    
```

```python
# Q.3) Write a program to convert a string to title case without using the title() method. 
# Example: "python programming" should become "Python Programming" 
s = input("Enter a string: ")
words = s.split()
result = ""

for w in words:
    first = w[0].upper()
    rest = w[1:].lower()
    result = result + first + rest + " "

print("Title case string:", result.strip())
```

```python
# Q.4) Write a program to replace all occurrences of a substring with another substring in a given string
s = input("Enter string: ")
old = input("Enter word to replace: ")
new = input("Enter new word: ")

s = s.replace(old, new)

print("Result:", s)
```

```python
# Q.1) Write a program to find the longest word in a sentence.

s = input("Enter a sentence: ")
words = s.split()
longest = words[0]

for ch in words:
    if len(ch) > len(longest):
        longest = ch

print("Longest word is:", longest)
```

```python
# Q.2) Write a program to check if a string contains only digits, only alphabets, or both.

s = input("Enter a string: ")
has_alpha = False
has_digit = False

for ch in s:
    if ch.isalpha():
        has_alpha = True
    elif ch.isdigit():
        has_digit = True

if has_alpha and has_digit:
    print("Contains both, alphabets and digits")
elif has_alpha:
    print("Contains only alphabets")
elif has_digit:
    print("Contains only digits")
else:
    print("Contains neither alphabets nor digits")
```
