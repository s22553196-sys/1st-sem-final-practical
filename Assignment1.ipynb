```python
# BASIC CALCULATIONS
a,b=2,5
print("Addition:", a+b)
print("subtraction:", a-b)
print("multiplication:", a*b)
print("division:", a/b)
print("float division:", a//b)
print("modulus:",a%b)
print("power:",a**b)
```

```python
# POWER OF A NUMBER
n=4
print("squre:",n**2)
print("cube:",n**3)
print("fifth power:",n**5)
```

```python
# APPLE DISTRIbUTION
apples,children=125,12
print("each child gets:",apples//children)
print("leftover apples:",apples%children)
     
```
