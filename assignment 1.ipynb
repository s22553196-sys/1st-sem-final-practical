```python
# Q.2) Write a program which accepts 6 integer values and prints "DUPLICATES" if any of the values entered are duplicates otherwise it prints "ALL UNIQUE".
# Read 6 integer values from the user
nums = []
for i in range(6):
    value = int(input("Enter an integer: "))
    nums.append(value)

if len(nums) != len(set(nums)):
    print("DUPLICATES")
else:
    print("ALL UNIQUE")
```

```python
# Q.4) Write a program which finds sum of digits of a number. 
# Example: n=130 then output is 4 (1+3+0).

n = int(input("Enter a number: "))
sum_digits = 0

while n > 0:
    digit = n % 10          
    sum_digits += digit    
    n = n // 10             

print("Sum of digits is:", sum_digits)
```

```python
# Q.5) Write a program which prints Fibonacci series of a number.

n = int(input("Enter number of terms: "))
a = 0
b = 1

print("Fibonacci series:")
for i in range(n):
    print(a, end=" ")
    c = a + b 
    a = b
    b = c
```

```python
# Q.1) Write a program which accept an integer value 'n' and display all prime numbers till 'n'.
n = int(input("Enter a number: "))
print("Prime numbers up to", n, "are:")

for num in range(2, n + 1):
    is_prime = True
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num, end=" ")
        
```

```python
#Q.4) Write a program to reverse a given number.
num = int(input("Enter a number: "))
reverse_num = 0

while num > 0:
    digit = num % 10             # Get last digit
    reverse_num = reverse_num * 10 + digit
    num = num // 10              # Remove last digit

print("Reversed number is:", reverse_num)
```

```python
# Q.1) Write a Sequential search function which searches an item in a sorted list. 
# The function should return the index of element to be searched in the list.
def sequential_search(arr, key):
    
    for index in range(len(arr)):
        if arr[index] == key:
            return index    
    return -1               

numbers = [5, 10, 15, 20, 25, 30]
search_item = 20
result = sequential_search(numbers, search_item)

if result != -1:
    print("Element found at index:", result)
else:
    print("Element not found")
    
```
